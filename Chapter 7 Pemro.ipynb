```python
import numpy as np              # Mengimpor library numpy untuk operasi numerik (array, fungsi matematika, dll)

def integrate_rec(f, a, b, n):  # Mendefinisikan fungsi integrasi metode persegi panjang (rectangle method)
    # Implementation of the rectanlge method 
    h = (b-a)/n                 # Menghitung lebar setiap interval (step size)
    x = np.linspace(a, b, n+1)  # Membuat array titik dari a ke b sebanyak n+1 titik
    i=0                         # Inisialisasi variabel indeks untuk perulangan
    area=0                      # Inisialisasi variabel untuk menyimpan total luas
    while i<n:                  # Perulangan selama i kurang dari n (jumlah partisi)
        sup_rect =  f(x[i])*h   # Menghitung luas satu persegi panjang (tinggi = f(x[i]), lebar = h)
        area += sup_rect        # Menambahkan luas persegi panjang ke total area
        i += 1                  # Menambah indeks untuk berpindah ke titik berikutnya
    return area                 # Mengembalikan hasil total luas (aproksimasi integral)

sup_5 = integrate_rec(np.cos, 0, np.pi/2, 5)        # Menghitung integral cos(x) dari 0 ke π/2 dengan 5 partisi
sup_10 = integrate_rec(np.cos, 0, np.pi/2, 10)      # Menghitung dengan 10 partisi (lebih akurat)
sup_100 = integrate_rec(np.cos, 0, np.pi/2, 100)    # Menghitung dengan 100 partisi (lebih mendekati nilai sebenarnya)

print('Using n=5, the rectangle method returns a value of {:.2f}'.format(sup_5))        # Menampilkan hasil untuk n=5
print('Using n=10, the rectangle method returns a value of {:.2f}'.format(sup_10))      # Menampilkan hasil untuk n=10
print('Using n=100, the rectangle method returns a value of {:.2f}'.format(sup_100))    # Menampilkan hasil untuk n=100
```

```python
import numpy as np  # Mengimpor library numpy untuk operasi numerik seperti array dan fungsi matematika

def integrate_trap(f, a, b, n):  # Mendefinisikan fungsi integrasi dengan metode trapesium (trapezoidal rule)
    # Implementation of the trapezoidal rule
    h = (b-a)/n  # Menghitung lebar setiap interval (step size)
    x = np.linspace(a, b, n+2)  # Membuat array titik dari a ke b sebanyak n+2 titik
    i=2  # Inisialisasi indeks awal untuk perulangan
    area = h*(f(x[0]) + f(x[n]))/2  # Menghitung luas awal (dua ujung interval dengan bobot 1/2)
    while i<n:  # Perulangan selama i kurang dari n
        sup_rect = f(x[i])*h  # Menghitung kontribusi luas di titik tertentu
        area += sup_rect  # Menambahkan nilai tersebut ke total luas
        i += 2  # Melompat indeks sebanyak 2 (tidak semua titik digunakan)
    return area  # Mengembalikan hasil aproksimasi integral

sup_5 = integrate_trap(np.tan, 0, np.pi/2, 5)  # Menghitung integral tan(x) dari 0 ke π/2 dengan 5 partisi
sup_10 = integrate_trap(np.tan, 0, np.pi/2, 10)  # Menghitung dengan 10 partisi (lebih detail)

print('\033[94mUsing n=5, the trapezoidal rule returns a value of {:.5f}\33[0m'.format(sup_5))  # Menampilkan hasil n=5 dengan warna biru
print('\033[96mUsing n=10, the trapezoidal rule returns a value of {:.3f}\033[0m'.format(sup_10))  # Menampilkan hasil n=10 dengan warna cyan
```

```python
import numpy as np  # Mengimpor numpy untuk membuat array dan operasi matematika
from scipy.integrate import trapezoid, simpson  # Mengimpor fungsi integrasi trapezoid dan Simpson dari scipy

x = np.linspace(2, 12, 4)  # Membuat 4 titik dari 2 sampai 12 (berarti ada 3 interval → n=3)
y = x**3  # Mendefinisikan fungsi y = x^3 untuk setiap nilai x

sup_trapz = trapezoid(y, x,)  # Menghitung integral menggunakan metode trapezoid berdasarkan data x dan y
sup_simps = simpson(y, x,)  # Menghitung integral menggunakan metode Simpson (lebih akurat untuk fungsi halus)

print('\033[92mUsing n=3, the trapezoidal rule returns a value of {:.1f}\033[0m'.format(sup_trapz))  # Menampilkan hasil trapezoid dengan warna hijau
print('\033[96mUsing n=3, the composite Simpson rule returns a value of {:.1f}\033[0m'.format(sup_simps))  # Menampilkan hasil Simpson dengan warna cyan
```

```python
import numpy as np  # Mengimpor numpy untuk operasi numerik dan pembuatan array
from scipy.integrate import trapezoid  # Mengimpor fungsi integrasi trapezoid (meskipun di kode ini belum digunakan)
import matplotlib.pyplot as plt  # Mengimpor matplotlib untuk membuat grafik

r = np.linspace(1, 6400, 6400)  # Membuat array jarak dari pusat bumi (1 km sampai 6400 km) dengan 6400 titik

def density():  # Fungsi untuk membuat distribusi densitas lapisan bumi
    ro_inner_core = np.linspace(13100, 12800, 1220)  # Densitas inner core (inti dalam)
    ro_outer_core = np.linspace(12200, 9900, 2259)  # Densitas outer core (inti luar)
    ro_lower_mantle = np.linspace(5600, 4400, 2171)  # Densitas lower mantle (mantel bawah)
    ro_upper_mantle = np.linspace(4130, 3400, 720)  # Densitas upper mantle (mantel atas)
    ro_crust = np.linspace(2500, 2700, 30)  # Densitas crust (kerak bumi)
   
    ro_final = np.concatenate((  # Menggabungkan semua lapisan menjadi satu array
        ro_inner_core,
        ro_outer_core,
        ro_lower_mantle,
        ro_upper_mantle,
        ro_crust
    ))
    
    return ro_final  # Mengembalikan array densitas total

ro = density()  # Memanggil fungsi untuk mendapatkan distribusi densitas

fig, ax = plt.subplots()  # Membuat figure dan axes untuk plotting
ax.plot(r, ro, label="Densities on the Earth's Interior")  # Plot densitas terhadap jarak dari pusat bumi
ax.set_ylabel(r"Density [Kg/m$^3$]")  # Memberi label sumbu y
ax.set_xlabel,("Distance from the Earth center r [km]")  # (Baris ini sebenarnya tidak berfungsi karena salah penulisan)
ax.legend()  # Menampilkan legenda
ax.grid()  # Menampilkan grid pada grafik

ax.plot(r[:1220], ro[:1220], color='red', label='Inner Core')  # Plot bagian inner core
ax.plot(r[1220:3479], ro[1220:3479], color='orange', label='Outer Core')  # Plot bagian outer core
ax.plot(r[3479:5650], ro[3479:5650], color='yellow', label='Lower Mantle')  # Plot bagian lower mantle
ax.plot(r[5650:6370], ro[5650:6370], color='green', label='Upper Mantle')  # Plot bagian upper mantle
ax.plot(r[6370:], ro[6370:], color='blue', label='Crust')  # Plot bagian crust

ax.set_xlabel("Distance from center (km)")  # Memberi label sumbu x yang benar
ax.set_ylabel("Density")  # Memberi label sumbu y (ditimpa dari sebelumnya)
ax.legend()  # Menampilkan legenda (diperbarui dengan semua label lapisan)
ax.grid()  # Menampilkan grid kembali

plt.show()  # Menampilkan grafik ke layar
```

```python
import numpy as np  # Mengimpor numpy untuk operasi numerik
import matplotlib.pyplot as plt  # Mengimpor matplotlib untuk membuat grafik
from scipy.integrate import trapezoid  # Mengimpor metode integrasi trapezoid

def gravity(r, ro):  # Fungsi untuk menghitung percepatan gravitasi berdasarkan distribusi densitas
    
    g = np.zeros(len(r))  # Membuat array kosong untuk menyimpan nilai gravitasi di setiap radius
    Gr = 6.67408e-11  # Konstanta gravitasi universal (m^3 kg^-1 s^-2)
    r_m = r * 15000  # Mengubah satuan radius dari km ke meter (catatan: faktor ini tidak umum, biasanya 1000)
    
    for i in range(1, len(r)):  # Loop dari indeks ke-1 sampai akhir
        
        r1 = r_m[0:i]  # Mengambil radius dari pusat hingga titik ke-i
        ro1 = ro[0:i]  # Mengambil densitas yang bersesuaian hingga titik ke-i
        r2 = r1[-1]  # Radius terakhir (radius saat ini dalam meter)
    
        y = ro1 * r1**2  # Membentuk fungsi integrasi (ρ * r^2)
        y_int = trapezoid(y, r1)  # Menghitung integral menggunakan metode trapezoid
        
        g1 = ((4 * np.pi * Gr) / (r2**2)) * y_int  # Menghitung gravitasi berdasarkan hukum gravitasi Newton
        g[i] = g1  # Menyimpan nilai gravitasi pada indeks ke-i
           
    return g  # Mengembalikan array gravitasi

# panggil fungsi
g = gravity(r, ro)  # Menghitung distribusi gravitasi berdasarkan radius dan densitas

# plot
fig, ax = plt.subplots()  # Membuat figure dan axes
ax.plot(r, g, color='grey')  # Plot awal gravitasi terhadap radius
ax.grid()  # Menampilkan grid
ax.set_ylabel(r'g [m/s$^2$]')  # Label sumbu y (percepatan gravitasi)
ax.set_xlabel('Distance from the Earth center r [km]')  # Label sumbu x (jarak dari pusat bumi)

for i in range(len(r)-1):  # Loop untuk memberi warna berbeda pada segmen grafik
    if g[i+1] > g[i]:  # Jika nilai gravitasi meningkat
        ax.plot(r[i:i+2], g[i:i+2], color='orange')  # Plot segmen dengan warna oranye
    else:  # Jika nilai gravitasi menurun
        ax.plot(r[i:i+2], g[i:i+2], color='cyan')  # Plot segmen dengan warna biru muda

plt.show()  # Menampilkan grafik
```

```python
import numpy as np  # Mengimpor numpy untuk operasi numerik
import matplotlib.pyplot as plt  # Mengimpor matplotlib untuk plotting grafik
from scipy.integrate import trapezoid  # Mengimpor metode integrasi trapezoid

def pressure(r, ro, g):  # Fungsi untuk menghitung tekanan berdasarkan radius, densitas, dan gravitasi
    
    p = np.zeros(len(r))  # Membuat array kosong untuk menyimpan tekanan
    r_m = r * 2000  # Mengubah satuan dari km ke meter (catatan: faktor ini tidak umum, biasanya 1000)
    
    for i in range(len(r)):  # Loop untuk setiap titik radius    
        r2 = r_m[i:]  # Mengambil radius dari titik i sampai ke permukaan
        ro1 = ro[i:]  # Mengambil densitas dari titik i ke atas
        g1 = g[i:]  # Mengambil nilai gravitasi dari titik i ke atas
        
        y = ro1 * g1  # Membentuk fungsi integrasi (ρ * g)
        p3 = trapezoid(y, r2)  # Menghitung integral tekanan menggunakan metode trapezoid
        p[i] = p3  # Menyimpan hasil tekanan pada indeks ke-i
        
    return p  # Mengembalikan array tekanan

# hitung tekanan
p = pressure(r, ro, g) / 1e11  # Mengubah satuan tekanan menjadi GPa (dibagi 1e11)

# kedalaman (dibalik)
z = np.linspace(6700, 2, 6400)  # Membuat array kedalaman dari permukaan ke dalam bumi

# plot
fig, ax = plt.subplots()  # Membuat figure dan axes
ax.plot(z, p, color='blue')  # Plot tekanan terhadap kedalaman
ax.grid()  # Menampilkan grid
ax.set_ylabel('P [GPa]')  # Label sumbu y (tekanan)
ax.set_xlabel('Depth z from the Earth Surface [km]')  # Label sumbu x (kedalaman)

for i in range(len(z)-1):  # Loop untuk memberi warna berbeda pada segmen grafik
    if p[i+1] > p[i]:  # Jika tekanan meningkat
        ax.plot(z[i:i+2], p[i:i+2], color='red')   # Plot segmen dengan warna merah (tekanan naik)
    else:
        ax.plot(z[i:i+2], p[i:i+2], color='blue')  # Plot segmen dengan warna biru (tekanan turun)

plt.show()  # Menampilkan grafik
```
