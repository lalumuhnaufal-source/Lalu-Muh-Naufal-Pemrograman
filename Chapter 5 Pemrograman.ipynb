```python
import pandas as pd
from scipy.stats.mstats import gmean, hmean
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

a_mean = my_dataset.Zr.mean()
g_mean = gmean(my_dataset['Zr'])
h_mean = hmean(my_dataset['Zr'])

print ('-------')
print ('arithmetic mean')
print ("{0:.1f} [ppm]".format(a_mean))
print ('-------')

print ('geometric mean')
print ("{0:.1f} [ppm]".format(g_mean))
print ('-------')

print ('harmonic mean')
print ("{0:.1f} [ppm]".format(h_mean))
print ('-------')

fig, ax = plt.subplots()
ax.hist(my_dataset.Zr, bins='auto', density=False, edgecolor='k', color='green', label='Measurements Hist', alpha=0.8) 
ax.axvline(a_mean, color='blue', label='Arithmetic mean', linewidth=2.5)
ax.axvline(g_mean, color='black',  label='Geometric mean', linewidth=2.5)
ax.axvline(h_mean, color='cyan',  label='Harmonic mean', linewidth=2.5)
ax.set_xlabel('Zr [ppm]')
ax.set_ylabel('Probability density')
ax.legend()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

hist, bin_edges = np.histogram(my_dataset['Zr'], bins= 30, density=True)
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+2])/3

print ('modal value: {0:.01f} [ppm]'.format(modal_value))

fig, ax = plt.subplots()
ax.hist(my_dataset.Zr, bins=30, density=False, edgecolor='k',color='black', label="Measurements Hist", alpha=0.8) 
ax.axvline(modal_value, color="orange", label="Modal value", linewidth=2)
ax.set_xlabel('Zr [ppm]')
ax.set_ylabel('Diagram Batang')
ax.legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

variance =  my_dataset['Nb'].var()
stddev =  my_dataset['Nb'].std()

print ('-------')
print ('Variance')
print("{0:.2f} [square ppm]".format(variance))
print ('-------')
print ('Standard Deviation')
print("{0:.2f} [ppm]".format(stddev))
print ('-------')

fig, ax = plt.subplots()
ax.hist(my_dataset.Nb, bins= 20, density = True, edgecolor='k',color='grey', label='Measurements Hist') 
ax.axvline(my_dataset['Nb'].mean()-stddev, color='purple', label=r'mean - 1$\sigma$', linewidth=2)
ax.axvline(my_dataset['Nb'].mean()+stddev, color='green', label=r'mean + 1$\sigma$', linewidth=2)
ax.axvspan(my_dataset['Nb'].mean()-stddev, my_dataset['Nb'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$')
ax.set_xlabel('Nb [ppm]')
ax.set_ylabel('Probability density')
ax.legend()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

q1 = np.percentile(my_dataset.Nb, 25, method= 'midpoint') 
q3 = np.percentile(my_dataset.Nb, 75, method= 'midpoint') 
  
iqr = q3 - q1 # Interquaritle range (IQR)  

print ('-------')
print ('method range (IQR): {0:.2f} [ppm]'.format(iqr))
print ('-------')

fig, ax = plt.subplots()
ax.hist(my_dataset.Nb, bins='auto', density=True, edgecolor='k',color='cyan', label='Measurements Hist') 
ax.axvline(q1, color='purple', label='Q1', linewidth=2)
ax.axvline(q3, color='green', label='Q3', linewidth=2)
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)')
ax.set_xlabel('Nb [ppm]')
ax.set_ylabel('Probability density')
ax.legend()
```

```python
import numpy as np
from scipy.stats import skew

a_mean = my_dataset.Nb.mean()
median = my_dataset.Nb.median()
hist, bin_edges = np.histogram(my_dataset['Nb'], bins=20, density=True)
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2
standard_deviation = my_dataset['Nb'].std()

a1 = (a_mean - modal_value) / standard_deviation 
a2 = 3 * (a_mean - median) / standard_deviation
g1 = skew(my_dataset['Nb'])

print ('-------')
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1))
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2))
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))
print ('-------')
```
