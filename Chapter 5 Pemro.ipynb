```python
import pandas as pd  # Mengimpor pandas untuk membaca dan mengolah data (file Excel)
from scipy.stats.mstats import gmean, hmean  # Mengimpor geometric mean dan harmonic mean
import matplotlib.pyplot as plt  # Mengimpor matplotlib untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # Membaca data dari file Excel pada sheet 'Supp_traces'

a_mean = my_dataset.Zr.mean()  # Menghitung rata-rata aritmatika (mean biasa) dari kolom Zr
g_mean = gmean(my_dataset['Zr'])  # Menghitung rata-rata geometri dari kolom Zr
h_mean = hmean(my_dataset['Zr'])  # Menghitung rata-rata harmonik dari kolom Zr

print ('-------')  # Garis pemisah output
print ('arithmetic mean')  # Judul untuk mean aritmatika
print ("{0:.1f} [ppm]".format(a_mean))  # Menampilkan nilai mean aritmatika dengan 1 angka desimal
print ('-------')  # Garis pemisah

print ('geometric mean')  # Judul untuk mean geometri
print ("{0:.1f} [ppm]".format(g_mean))  # Menampilkan nilai mean geometri
print ('-------')  # Garis pemisah

print ('harmonic mean')  # Judul untuk mean harmonik
print ("{0:.1f} [ppm]".format(h_mean))  # Menampilkan nilai mean harmonik
print ('-------')  # Garis pemisah

fig, ax = plt.subplots()  # Membuat figure dan axes untuk plotting
ax.hist(my_dataset.Zr, bins='auto', density=False, edgecolor='k', color='green', label='Measurements Hist', alpha=0.8)  # Membuat histogram data Zr
ax.axvline(a_mean, color='blue', label='Arithmetic mean', linewidth=2.5)  # Menambahkan garis vertikal untuk mean aritmatika
ax.axvline(g_mean, color='black',  label='Geometric mean', linewidth=2.5)  # Menambahkan garis vertikal untuk mean geometri
ax.axvline(h_mean, color='cyan',  label='Harmonic mean', linewidth=2.5)  # Menambahkan garis vertikal untuk mean harmonik
ax.set_xlabel('Zr [ppm]')  # Memberi label sumbu x
ax.set_ylabel('Probability density')  # Memberi label sumbu y (meskipun sebenarnya ini frekuensi, bukan density)
ax.legend()  # Menampilkan legenda
```

```python
import pandas as pd  # Library untuk membaca dan mengolah data
import matplotlib.pyplot as plt  # Library untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # Membaca data dari file Excel

# Menghitung varians (penyebaran data terhadap rata-rata)
variance =  my_dataset['Nb'].var()

# Menghitung standar deviasi (akar dari varians)
stddev =  my_dataset['Nb'].std()

print ('-------')  # Garis pemisah output
print ('Variance')  # Judul output varians
print("{0:.2f} [square ppm]".format(variance))  # Menampilkan varians (satuan ppm^2)
print ('-------')

print ('Standard Deviation')  # Judul output standar deviasi
print("{0:.2f} [ppm]".format(stddev))  # Menampilkan standar deviasi
print ('-------')

fig, ax = plt.subplots()  # Membuat area grafik (figure dan axes)

# Membuat histogram data Nb
ax.hist(
    my_dataset.Nb,       # Data yang diplot
    bins=20,             # Jumlah kelas (bin)
    density=True,        # Menampilkan dalam bentuk probabilitas (density)
    edgecolor='k',       # Warna tepi batang (hitam)
    color='grey',        # Warna batang histogram
    label='Measurements Hist'  # Label legenda
) 

# Menambahkan garis untuk mean - 1 standar deviasi
ax.axvline(
    my_dataset['Nb'].mean()-stddev,  # Posisi mean - σ
    color='purple',                  # Warna garis
    label=r'mean - 1$\sigma$',       # Label legenda (format LaTeX)
    linewidth=2
)

# Menambahkan garis untuk mean + 1 standar deviasi
ax.axvline(
    my_dataset['Nb'].mean()+stddev,  # Posisi mean + σ
    color='green',                   # Warna garis
    label=r'mean + 1$\sigma$',       # Label legenda
    linewidth=2
)

# Menambahkan area bayangan antara mean ± 1σ
ax.axvspan(
    my_dataset['Nb'].mean()-stddev,  # Batas kiri
    my_dataset['Nb'].mean()+stddev,  # Batas kanan
    alpha=0.1,                       # Transparansi area
    color='orange',                  # Warna area
    label=r'mean $\pm$ 1$\sigma$'    # Label legenda
)

ax.set_xlabel('Nb [ppm]')  # Label sumbu x (konsentrasi Nb)
ax.set_ylabel('Probability density')  # Label sumbu y (kepadatan probabilitas)
ax.legend()  # Menampilkan legenda
```

```python
import pandas as pd  # Library untuk membaca dan mengolah data
import numpy as np  # Library untuk operasi numerik (termasuk perhitungan persentil)
import matplotlib.pyplot as plt  # Library untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # Membaca data dari file Excel

# Menghitung kuartil pertama (Q1 = 25th percentile)
q1 = np.percentile(my_dataset.Nb, 25, method='midpoint') 

# Menghitung kuartil ketiga (Q3 = 75th percentile)
q3 = np.percentile(my_dataset.Nb, 75, method='midpoint') 
  
# Menghitung Interquartile Range (IQR)
iqr = q3 - q1  # IQR = Q3 - Q1 (mengukur sebaran data bagian tengah)

print ('-------')  # Garis pemisah output
print ('method range (IQR): {0:.2f} [ppm]'.format(iqr))  # Menampilkan nilai IQR
print ('-------')

fig, ax = plt.subplots()  # Membuat area grafik (figure dan axes)

# Membuat histogram data Nb
ax.hist(
    my_dataset.Nb,         # Data yang diplot
    bins='auto',           # Jumlah bin otomatis
    density=True,          # Menampilkan dalam bentuk probabilitas
    edgecolor='k',         # Warna tepi batang (hitam)
    color='cyan',          # Warna batang histogram
    label='Measurements Hist'  # Label legenda
) 

# Menambahkan garis vertikal untuk Q1
ax.axvline(
    q1,                    # Posisi Q1
    color='purple',        # Warna garis
    label='Q1',            # Label legenda
    linewidth=2
)

# Menambahkan garis vertikal untuk Q3
ax.axvline(
    q3,                    # Posisi Q3
    color='green',         # Warna garis
    label='Q3',            # Label legenda
    linewidth=2
)

# Menambahkan area bayangan untuk IQR
ax.axvspan(
    q1,                    # Batas kiri (Q1)
    q3,                    # Batas kanan (Q3)
    alpha=0.1,             # Transparansi area
    color='orange',        # Warna area
    label='Interquaritle range (IQR)'  # Label legenda
)

ax.set_xlabel('Nb [ppm]')  # Label sumbu x (konsentrasi Nb)
ax.set_ylabel('Probability density')  # Label sumbu y
ax.legend()  # Menampilkan legenda
```

```python
import pandas as pd  # Library untuk membaca dan mengolah data
import numpy as np  # Library untuk operasi numerik
import matplotlib.pyplot as plt  # Library untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # Membaca data dari file Excel

# Menghitung mean (rata-rata aritmatika)
a_mean = my_dataset.La.mean()

# Menghitung median (nilai tengah data)
median = my_dataset.La.median()

# Menghitung histogram untuk mendapatkan modus
hist, bin_edges = np.histogram(my_dataset['La'], bins=20, density=True)

# Menghitung nilai modus (berdasarkan bin dengan frekuensi tertinggi)
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1]) / 2

fig, ax = plt.subplots()  # Membuat area grafik

# Membuat histogram data La
ax.hist(
    my_dataset.La,        # Data yang diplot
    bins=20,              # Jumlah kelas (bin)
    density=True,         # Menampilkan dalam bentuk probabilitas
    edgecolor='k',        # Warna tepi batang
    color='blue',         # Warna batang histogram
    label="Measurements Hist"  # Label legenda
) 

# Menambahkan garis vertikal untuk modus
ax.axvline(
    modal_value,          # Posisi modus
    color='orange',       # Warna garis
    label='Modal Value',  # Label legenda
    linewidth=2
)

# Menambahkan garis vertikal untuk median
ax.axvline(
    median,               # Posisi median
    color='purple',       # Warna garis
    label='Median Value', # Label legenda
    linewidth=2
)

# Menambahkan garis vertikal untuk mean
ax.axvline(
    a_mean,               # Posisi mean
    color='green',        # Warna garis
    label='Arithmetic mean',  # Label legenda
    linewidth=2
)

ax.set_xlabel('La [ppm]')  # Label sumbu x (konsentrasi La)
ax.set_ylabel('Probability density')  # Label sumbu y (kepadatan probabilitas)
ax.legend()  # Menampilkan legenda
```

```python
import numpy as np  # Library untuk operasi numerik
from scipy.stats import skew  # Mengimpor fungsi untuk menghitung skewness (kemencengan distribusi)

# Menghitung mean (rata-rata aritmatika)
a_mean = my_dataset.Nb.mean()

# Menghitung median (nilai tengah)
median = my_dataset.Nb.median()

# Membuat histogram untuk mendapatkan modus
hist, bin_edges = np.histogram(my_dataset['Nb'], bins=20, density=True)

# Menghitung nilai modus (berdasarkan bin dengan frekuensi tertinggi)
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1]) / 2

# Menghitung standar deviasi (ukuran penyebaran data)
standard_deviation = my_dataset['Nb'].std()

# Pearson’s first coefficient of skewness (berdasarkan mean dan modus)
a1 = (a_mean - modal_value) / standard_deviation 

# Pearson’s second coefficient of skewness (berdasarkan mean dan median)
a2 = 3 * (a_mean - median) / standard_deviation

# Fisher-Pearson coefficient of skewness (menggunakan fungsi bawaan scipy)
g1 = skew(my_dataset['Nb'])

print ('-------')  # Garis pemisah output

# Menampilkan hasil koefisien skewness
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1))
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2))
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))

print ('-------')  # Garis pemisah
```
